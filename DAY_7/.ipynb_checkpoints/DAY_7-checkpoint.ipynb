# IRIS FLOWER DETECTION

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data,columns = iris.feature_names)
print(df)
```

#### basic info about data

```python
df["species"]  = iris.target
df["species"] = df["species"].map({0:"setosa",1:"versicolor",2:"verginica"})
```

```python
print("shape of data",df.shape)
print("first 5 rows",df.head())
```

```python
print("datatypes", df.dtypes)
```

```python
print("null values", df.isnull().sum())
```

#### summary stat

```python
print("summary stat", df.describe())
```

#### class distribution

```python
print("species distribution", df["species"].value_counts())
```

#### co relation matrix

```python
print("correlation matrix:\n", df.select_dtypes(include='number').corr())
```

#### pair plot(seaborn)

```python
sb.pairplot(df, hue = "species", diag_kind = "kde")
plt.suptitle("pairplot - iris data set", y=1.02)
plt.show()
```

#### box plot for each feature

```python
plt.figure(figsize =(12,6))
for i ,col in enumerate(df.columns[:-1]):
    plt.subplot(2,2,i+1)
    sb.boxplot(x = "species", y = col, data = df)
    plt.title(f"{col} by species")
plt.tight_layout()
plt.show()    
```

#### co relation heatmap

```python
plt.figure(figsize =(8,6))
sb.heatmap(df.select_dtypes(include='number').corr(), annot = True, cmap = "coolwarm")
plt.title("feature corelation heatmap")
plt.show()
```
