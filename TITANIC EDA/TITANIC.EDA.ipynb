```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
titanic = sb.load_dataset("titanic")
```

```python
print(titanic)
```

```python
print("datatypes",titanic.dtypes)
```

#### to print the sum of null values 

```python
print("null values", titanic.isnull().sum())
```

#### summary stat

```python
print("summary",titanic.describe(include = "all"))
```

#### co relation matrix

```python
print("correlation matrix:\n", titanic.select_dtypes(include='number').corr())
```

#### drop columns with too many missing values

```python
titanic = titanic.drop(["deck"],axis = 1)
print(titanic)
```

#### converting survived column from 1 - 0 to survived - not-survived

```python
titanic["survived"] = titanic["survived"].map({0:"not-survived" , 1:"survived"})
print(titanic)
```

#### count of survived vs not survived

```python
sb.countplot(x="survived",data = titanic)
plt.title("survival count")
plt.xlabel("survived (0 = no, 1 = yes)")
plt.ylabel("count")
plt.show()
```

#### survival by sex

```python
sb.countplot(x="sex",data = titanic)
plt.title("survival count by sex")
plt.xlabel("survived male , female")
plt.ylabel("count")
plt.show()
```

```python
sb.countplot(x="sex", hue = "survived", data = titanic)
plt.title("survival count by sex")
plt.xlabel("survived male , female")
plt.ylabel("count")
plt.show()
```

#### survival by class

```python
sb.countplot(x = "pclass", hue = "survived", data = titanic)
plt.title("passenger class wise survived ")
plt.show()
```

#### town wise survival

```python
sb.countplot(x = "embark_town", hue = "survived", data = titanic)
plt.title("town wise survival ")
plt.show()
```

#### age distribution

```python
plt.hist(titanic["age"].dropna(), bins = 50, color = "red", edgecolor = "black")
plt.title("age distribution")
plt.xlabel("Age")
plt.ylabel("frequency")
plt.show()
```

#### age by survival

```python
sb.boxplot(x = "survived" , y = "age" , data = titanic)
plt.title("age distribution by survival")
plt.xlabel("survival status")
plt.ylabel("Age")
plt.show()
```

#### co relation heatmap

```python
plt.figure(figsize = (10,6))
sb.heatmap(titanic.corr(numeric_only=True),annot = True, cmap = "coolwarm")
plt.title("co relation heatmap")
plt.show()
print(titanic.corr(numeric_only=True))
```
