# Model Evaluation Metrics (confusion matrix, precision, recall, F1)

#### why metrics matter?
- imagine a model that predicts desease
- predict desies
- you care not just about overall accuracy
- you want to know:
- how many sick people are caught(recall)
- how often the model gives correct alerts(precision)
- how balance the performance is(F1 score)

#### Key Metrics:
| Metric | Meaning |
|--------|---------|
| accuracy | percentage of total correct prediction |
| precision | out of predicted positives, how many correct|
| recall | out of actual positives, how many detected |
| F1 score | balance between precision and recall(good when classes are imbalanced) | 

#### visual confusion matrix:

|  | Predicted positive | /predicted negative | 
|--|--------------------|---------------------|
| Actual positive | true positive(tp) | false negative(fn) |
| Actual negative | false positive(fn) | true negative(tn) |


#### Formulas: 
1.accuracy = (tp+tn)/total 

2.precision = tp/(tp+fp)

3.recall = tp/(tp+fn)

4.F1-score = 2*(precision*recall)/(precision + recall)

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score
```

```python
data = load_iris()
df = pd.DataFrame(data.data, columns = data.feature_names)
df["species"] = data.target
print(df)
```

```python
X = df.drop("species", axis = 1)
y = df["species"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 80, random_state = 42)
```

#### KNN model

```python
KNN = KNeighborsClassifier(n_neighbors = 5)
KNN.fit(X_train,y_train)
```

```python
y_pred = KNN.predict(X_test)
```

#### confusion matrix

```python
print(confusion_matrix(y_pred,y_test))
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True, cmap = "Blues", fmt = "d")
plt.title("confusion matrix")
plt.xlabel("predicted")
plt.ylabel("actual")
plt.show()
```

#### calculate metrics

```python
print(accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(precision_score(y_test, y_pred, average = "macro"))
```

```python
print(recall_score(y_test, y_pred, average = "macro"))
```

```python
print(f1_score(y_test, y_pred ,average = "macro"))
```
