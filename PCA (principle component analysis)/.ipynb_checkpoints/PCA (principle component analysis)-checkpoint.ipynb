# PCA (Principle Component Analysis) (Dimentonality Reduction)

#### what is `Dimentionality Reduction?`

- imagine you have 100 columns `(feature)` in your data
- some features are repeatative or not useful
- you want to keep the most important information but reduce columns

#### Real Life Example: 
- think of a huge textbook you summarize keypoints on one page
- `PCA` does the same reduces many features to fewer, important   

#### What is PCA?
- PCA simplifies your data by finding new hidden features called `principle components`
- these components capture the most variance `(information)`
- are combination of most features
- helps visualize complex data in `2-D` or `3-D` 

#### PCA is used for:
- visualizing high dimentional data
- speeding up machine learning models
- reducing noise `(incomplete data)`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
data = load_iris()
df = pd.DataFrame(data.data, columns = data.feature_names)
df["species"] = data.target
print(df)
```

#### Apply PCA to reduce features

```python
X = df.drop("species", axis = 1)
pca = PCA(n_components = 2)
components = pca.fit_transform(X)
```

```python
df["PC1"] = components[:, 0]
df["PC2"] = components[:, 1]
```

#### Visualize the PCA result

```python
plt.figure(figsize = (7,5))
sns.scatterplot(x = "PC1", y = "PC2", hue = "species", palette = "deep", data = df)
```

#### what happened here
- We had 4 features
- **sepal length** and **sepal width** got combined into **PC1**
- **petal length** and **petal width** got combined into **PC2**
- PCA combined them into 2 new features `Principle Component 1 (PC1)` & `Principle Component 2 (PC2)`
- these new features capture most of the important patterns in the data

#### Simple Analogy

- imagine a 3-D object like a cube
- you take its 2-D shadow on a wall
- The shadow still shows the shape, but its simpler
- PCA works similarly, reduced dimensions but keeps patterns

## Key Points :

- PCA is unsupervised: it doesnt use labels during reduction
- Helps with visualization and simplifying
- Principal Components are combinations of original features
- You decide how many components to keep `(n_components)`
