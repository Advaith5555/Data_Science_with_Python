# INTRODUCTION TO MATPLOTLIB

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Basic Line 

```python
x = np.linspace(0,10)
y = np.linspace(0,8)

plt.plot(x,y)
plt.title("sine wave")
plt.xlabel("x lable")
plt.ylabel("this is y lable")
plt.show()
```

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1,label="sine",color="red",linestyle="--")
plt.plot(x,y2,label="cosine",color="green",linestyle="dotted")

plt.title("sine wave & cosine wave")
plt.xlabel("x lable")
plt.ylabel("this is y lable")
plt.legend()
plt.grid(True)
plt.show()
```

#### multiple subplots

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

fig,axs = plt.subplots(2)
axs[0].plot(x,y1,color="green")
axs[0].set_title("sine")
axs[1].plot(x,y2,color="red")
axs[1].set_title("cosine")
plt.tight_layout()
plt.show()
```

#### bar chart

```python
fruits = ["apple","mango","banana","chiku"]
prices = [100,200,300,400]
plt.bar(fruits,prices,color="skyblue")
plt.title("prices of fruits")
plt.xlabel("fruits")
plt.ylabel("prices")
plt.show()
```

#### pie chart

```python
subjects = ["english","maths","history","science","geography","hindi","marathi"]
marks = [67,78,89,76,59,77,95]
plt.pie(marks, labels=subjects,autopct="%1.2f%%",startangle=180)
plt.title("subject wise marks")
plt.axis("equal")
plt.show()
```

#### histogram

```python
data = np.random.randn(10)
plt.hist(data,color="orange",edgecolor="green",bins=6)
plt.title("histogram")
plt.xlabel("this is x label")
plt.ylabel("this is y label")
plt.show()
```
