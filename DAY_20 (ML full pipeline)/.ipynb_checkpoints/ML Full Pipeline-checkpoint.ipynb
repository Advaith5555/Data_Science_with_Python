# ML Full Pipeline

### What is ml pipeline?

##### An ML pipeline is the complete process of building a machine learning model:
- Data loading
- Data cleanin & Preprocessing
- Exploratory Data Analysis (EDA)
- Feature Selection/Engineering
- Model building
- Model evaluation
- Making predictions

##### Think of it like cooking
- we collect ingredients => data loading
- clean them => data cleaning
- taste and explore flavours => EDA
- choose important ingredients => feature selection
- cook the dish => model building
- taste test => evaluation
- serve the dish => prediction

### 1. Import Libraries

### 4.Data Cleaning

```python
# drop columns with many null values
titanic.drop(['deck', 'embark_town'], axis=1, inplace=True)

# fill missing age with median
titanic['age'].fillna(titanic['age'].median(), inplace=True)

# drop remaining rows with nullvalues
titanic.dropna(inplace=True)
```
