# K-Means Clustering (implement & visualization)

#### what is `K-Means`?
- K-Means groups data into clusters based on similarity
- you tell the algorithm how many groups `(clusters)` you want `(k)`
- it finds the centers of those groups and assigns data points to the nearest one
- example: you have mixed fruits sizes => K-Means separates them into small, medium, large groups without being told in advance

#### what is clustering?
- clustering means `grouping similar things together`
- you dont tell the computer the groups it finds them `based on patterns` 
- real life example: imagine you have a basket of mixed fruit (apple, oranges, bananas) no labels. The computer loops at size, color, shape and groups similar fruits together

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
data = load_iris()
df = pd.DataFrame(data.data, columns = data.feature_names)
df["species"] = data.target
print(df)
```

```python
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', data=df)
plt.title("petal length vs width (Unlabeled)")
plt.show()
```

```python
X = df[["petal length (cm)","petal width (cm)"]]
kmeans = KMeans(n_clusters = 3, random_state = 42)
kmeans.fit(X)
```

```python
df["cluster"] = kmeans.labels_
```

```python
plt.figure(figsize = (8,6))
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', data=df, hue = "cluster", palette = "deep")
plt.title("petal length vs width (Unlabeled)")
plt.show()
```

#### plot cluster centers

```python
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], s = 200, c = "black", marker = "X", label = "Centers")
plt.title("kmeans clustering of samples(petal features)")
plt.legend()
plt.show()
```
