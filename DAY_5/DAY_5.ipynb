# PROBABILITY AND STAT

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
import seaborn as sns
sns.set(style = "whitegrid")
```

```python
marks = [65,70,75,80,70,85,90,95,70,100]
df = pd.DataFrame({"Marks":marks})
print(df)

```

#### mean

```python
mean_value = np.mean(marks)
print(mean_value)
```

#### meadian

```python
median_value = np.median(marks)
print(median_value)
```

#### mode

```python
mode = Counter(marks).most_common(1)[0][0]
print(mode)
```

#### variance

```python
variance_value = np.var(marks)
print(f"Variance value of {marks} is {variance_value}")
```

#### standard deviation

```python
stan_dev = np.std(marks)
print(stan_dev)
```

#### histogram wirh matplotlib

```python
plt.figure(figsize = (8,5))
plt.hist(marks,color = "orange",edgecolor = "black",bins=5)
plt.title("histogram")
plt.grid(True)
plt.xlabel("marks")
plt.ylabel("no of students")
plt.show()
```

#### hitogram with seaborn

```python
plt.figure(figsize = (8,5))
sns.histplot(df['Marks'],bins=5,kde = True,color = "green")
plt.title("histogram with kde")
plt.xlabel("marks")
plt.ylabel("density")
plt.show()
```

#### normal distribution

```python
normal_data = np.random.normal(loc = 50,scale = 10,size = 100)
normal_df = pd.DataFrame({"value":normal_data})

plt.figure(figsize = (8,5))
sns.histplot(normal_df['value'],bins=5,kde = True,color = "purple")
plt.title("normal distribution(mean = 50 & std = 10)")
plt.xlabel("value")
plt.ylabel("frequency")
plt.show()
```

```python
uniform_data = np.random.uniform(low = 10,high = 100,size = 1000)
uniform_df = pd.DataFrame({"value":uniform_data})

plt.figure(figsize = (8,5))
sns.histplot(uniform_df['value'],kde = True,color = "red")
plt.title("uniform distribution(10 - 100)")
plt.xlabel("value")
plt.ylabel("frequency")
plt.show()
```
